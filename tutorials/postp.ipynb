## process multiple runs

```python
runs = []
runs += ['jet_cfg1_wp75_4km_0a1500j']
runs += ['jet_cfg1_wp75_4km_1500a2000j_itide']
suffs = ['file_', ''] # this will disapear eventually
R = [pp.Run(root_path+run, prefix=s, open_nc=['ave']) for run, s in zip(runs,suffs)]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
for r, cfg in zip(R,runs):
    r.stats['KINETIC_ENRG'].plot(ax=ax, label=cfg)
ax.legend()
ax.grid()
```
