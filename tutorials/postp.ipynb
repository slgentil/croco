## process multiple runs

```python
runs = []
runs += ['jet_cfg1_wp6_4km_0a1000j']
runs += ['jet_cfg1_wp6_m1_4km_500a1000j']
R = [pp.Run(root_path+run, outputs=['ave']) for run in runs]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
for r, cfg in zip(R,runs):
    r.stats['KINETIC_ENRG'].plot(ax=ax, label=cfg)
ax.legend()
ax.grid()
```
